```python
raw_df = spark.read.format('binaryFile').load(base_path)
display(raw_df)
```

```python
from pyspark.sql.functions import expr

df_parsed = raw_df.select(
    "path",
    expr("ai_parse_document(content) as parsed_doc")
)

display(df_parsed.limit(3))
```

```python
elements_json_df = df_parsed.select(
    "path",
    expr("cast(variant_get(parsed_doc, '$.document.elements') as string) as elements_json")
)

display(elements_json_df)
```

```python
from pyspark.sql.functions import col, from_json, explode

elements_schema = """
array<struct<
 id:int,
 type:string,
 content:string,
 page_id:int,
 bbox:struct<coord: array<int>,page_id:int>,
 description:string
>>
"""

elements_df = (
    elements_json_df
    .select("path", explode(from_json(col("elements_json"), elements_schema)).alias("e"))
    .select(
        "path",
        col("e.id").alias("element_id"),
        col("e.type").alias("element_type"),
        col("e.content").alias("content"),
        col("e.page_id").alias("page_id"),
        col("e.bbox.coord").alias("bbox_coord"),
        col("e.description").alias("description")
    )
)

display(elements_df)
```

```python
from pyspark.sql import functions as F
from pyspark.sql.window import Window

# Step 1 — Define a window partitioned by each PDF file, ordered by element_id
windowSpec = Window.partitionBy("path").orderBy("element_id")

# Step 2 — Use lead() to fetch the content and type of the next row
elements_with_lead = elements_df \
    .withColumn("next_content",      F.lead("content", 1).over(windowSpec)) \
    .withColumn("next_element_type", F.lead("element_type", 1).over(windowSpec))

# Step 3 — Fix content based on the following cases:
# Case 1: If a row is a section_header with text "Total Claim Amount"
#         and the next row contains the amount (e.g. "Rs. X"),
#         merge both into a single row.
# Case 2: All other rows remain unchanged.
elements_df_fixed = elements_with_lead.withColumn(
    "content_fixed",
    F.when(
        (F.lower(F.col("content")).contains("total claim amount")) &
        (F.col("element_type") == "section_header"),
        # Combine: "Total Claim Amount" + " " + "Rs. 3,873,556.00"
        F.concat_ws(" ", F.col("content"), F.col("next_content"))
    )
    .otherwise(F.col("content"))
).withColumn(
    # Replace the original content column with the fixed content
    "content", F.col("content_fixed")
).drop("content_fixed", "next_content", "next_element_type")

# Step 4 — Remove the now-redundant rows
# These are rows that only contained the amount value (e.g. "Rs. X")
# and have already been merged into the previous row.
# Use lag() to look at the previous row's content and filter them out.
elements_df_clean = elements_df_fixed \
    .withColumn("prev_content", F.lag("content", 1).over(windowSpec)) \
    .filter(
        # Drop a row if the previous row contained "total claim amount"
        # and the current row is just a standalone amount value — it is now redundant.
        ~(
            F.lower(F.col("prev_content")).contains("total claim amount") &
            F.col("content").rlike(r"^Rs\.?\s*[\d,]+")
        )
    ) \
    .drop("prev_content")
```

```python
display(elements_df_clean)
```

```python
keywords = [
    # Common
    "total claim amount", "claim status",
    # AUTO
    "vehicle body repair", "spare parts", "towing", "surveyor",
    # HLTH
    "room charges", "operation theatre", "anaesthesia",
    "medicines", "diagnostic", "icu",
    # PROP
    "structural damage", "machinery", "stock", "electrical rewiring",
    "debris", "loss of business","loss assessment"
]
```

```python
from pyspark.sql.functions import col,lower

claim_amount_candidates = (
    elements_df_clean
    .withColumn("content_l", lower(col("content")))
    .where(" OR ".join([f"content_l LIKE '%{k}%'" for k in keywords])) 
    .select("path",  "element_type", "content","element_id")
    .orderBy("path", "element_id")
)

display(claim_amount_candidates)
```

```python
pairs_src = (
    claim_amount_candidates
    .withColumn("plain",
        F.regexp_replace(F.col("content"), r"<[^>]+>", " "))
    .withColumn("plain",
        F.regexp_replace(F.col("plain"), r"\s+", " "))
    .withColumn("plain_l",
        F.lower(F.col("plain")))
    .withColumn("claim_type",
        F.when(F.col("path").contains("AUTO"), "AUTO")
         .when(F.col("path").contains("HLTH"), "HLTH")
         .when(F.col("path").contains("PROP"), "PROP")
         .otherwise("UNKNOWN"))
)
```

```python
display(pairs_src)
```

```python
auto_level_df  =  pairs_src.filter(F.col("claim_type") == "AUTO")

auto_df = auto_level_df.select(
    "path",
    "claim_type",
    F.regexp_extract("plain",
        r"Vehicle Body Repair.*?Rs\.?\s*([\d,]+\.?\d{0,2})", 1
    ).alias("vehicle_repair"),

    F.regexp_extract("plain",
        r"Spare Parts.*?Rs\.?\s*([\d,]+\.?\d{0,2})", 1
    ).alias("spare_parts"),

    F.regexp_extract("plain",
        r"Towing.*?Rs\.?\s*([\d,]+\.?\d{0,2})", 1
    ).alias("towing_charges"),

    F.regexp_extract("plain",
        r"Surveyor.*?Rs\.?\s*([\d,]+\.?\d{0,2})", 1
    ).alias("survey_fee"),

    F.regexp_extract("plain",
        r"Total Claim Amount.*?Rs\.?\s*([\d,]+\.?\d{0,2})", 1
    ).alias("total_claim_amount"),

    # enum based regex
    F.regexp_extract("plain",
    r"(PENDING SURVEY|SURVEYOR APPOINTED|REJECTED|APPROVED|SETTLED|DOCUMENTS VERIFIED)", 1
    ).alias("claim_status"),
)

display(auto_df)
```

```python
merge_content = (
    pairs_src
    .groupBy("path", "claim_type")
    .agg(
        F.concat_ws(" ", F.collect_list("plain")).alias("full_text")
    )
)

display(merge_content)
```

```python

auto_level_df  =  merge_content.filter(F.col("claim_type") == "AUTO")

auto_df = auto_level_df.select(
    "path",
    "claim_type",
    F.regexp_extract("full_text",
        r"Vehicle Body Repair.*?Rs\.?\s*([\d,]+\.?\d{0,2})", 1
    ).alias("vehicle_repair"),

    F.regexp_extract("full_text",
        r"Spare Parts.*?Rs\.?\s*([\d,]+\.?\d{0,2})", 1
    ).alias("spare_parts"),

    F.regexp_extract("full_text",
        r"Towing.*?Rs\.?\s*([\d,]+\.?\d{0,2})", 1
    ).alias("towing_charges"),

    F.regexp_extract("full_text",
        r"Surveyor.*?Rs\.?\s*([\d,]+\.?\d{0,2})", 1
    ).alias("survey_fee"),

    F.regexp_extract("full_text",
        r"Total Claim Amount.*?Rs\.?\s*([\d,]+\.?\d{0,2})", 1
    ).alias("total_claim_amount"),

    # F.regexp_extract("full_text",
    # r"(?:Claim\s*Status|Status)\s*[:\-]?\s*([A-Za-z ]+)",1
    # ).alias("claim_status")

    # enum based regex
    F.regexp_extract("full_text",
    r"(PENDING SURVEY|SURVEYOR APPOINTED|REJECTED|APPROVED|SETTLED|DOCUMENTS VERIFIED)", 1
    ).alias("claim_status"),
)

display(auto_df)
```

```python
health_level_df  =  merge_content.filter(col("claim_type") == "HLTH")

health_df = health_level_df.select(
    "path",
    F.lit("HLTH").alias("claim_type"),
    # Room Charges (after closing bracket)
    F.regexp_extract(
        "full_text",
        r"Room Charges[^)]*\)\s*Rs\.?\s*([\d,]+\.?\d{0,2})",
        1
    ).alias("room_charges"),

    F.regexp_extract(
        "full_text",
        r"Operation Theatre.*?Rs\.?\s*([\d,]+\.?\d{0,2})",
        1
    ).alias("ot_fee"),

    F.regexp_extract(
        "full_text",
        r"Anaesthesia.*?Rs\.?\s*([\d,]+\.?\d{0,2})",
        1
    ).alias("anaesthesia"),

    F.regexp_extract(
        "full_text",
        r"Medicines.*?Rs\.?\s*([\d,]+\.?\d{0,2})",
        1
    ).alias("medicines"),

    F.regexp_extract(
        "full_text",
        r"Diagnostic.*?Rs\.?\s*([\d,]+\.?\d{0,2})",
        1
    ).alias("diagnostic_tests"),

    F.regexp_extract(
        "full_text",
        r"ICU[^)]*\)\s*Rs\.?\s*([\d,]+\.?\d{0,2})",
        1
    ).alias("icu_charges"),

    F.regexp_extract(
        "full_text",
        r"Total Claim Amount.*?Rs\.?\s*([\d,]+\.?\d{0,2})",
        1
    ).alias("total_claim_amount"),

    F.regexp_extract("full_text",
    r"(?:Claim\s*Status|Status)\s*[:\-]?\s*([A-Za-z ]+)",1
    ).alias("claim_status")
)

display(health_df)
```

```python
property_level_df  =  merge_content.filter(col("claim_type") == "PROP")

prop_df = property_level_df.select(
    "path",
    F.lit("PROP").alias("claim_type"),
    F.regexp_extract(
        "full_text",
        r"Structural Damage.*?Rs\.?\s*([\d,]+\.?\d{0,2})",
        1
    ).alias("structural_damage"),

    F.regexp_extract(
        "full_text",
        r"Machinery.*?Rs\.?\s*([\d,]+\.?\d{0,2})",
        1
    ).alias("machinery_damage"),

    F.regexp_extract(
        "full_text",
        r"Stock.*?Rs\.?\s*([\d,]+\.?\d{0,2})",
        1
    ).alias("stock_loss"),

    F.regexp_extract(
        "full_text",
        r"Electrical.*?Rs\.?\s*([\d,]+\.?\d{0,2})",
        1
    ).alias("electrical"),

    F.regexp_extract(
        "full_text",
        r"Debris.*?Rs\.?\s*([\d,]+\.?\d{0,2})",
        1
    ).alias("debris_removal"),

    F.regexp_extract(
        "full_text",
        r"Loss of Business.*?Rs\.?\s*([\d,]+\.?\d{0,2})",
        1
    ).alias("business_loss"),

    F.regexp_extract(
        "full_text",
        r"Total Claim Amount.*?Rs\.?\s*([\d,]+\.?\d{0,2})",
        1
    ).alias("total_claim_amount"),

    F.regexp_extract("full_text",
    r"(?:Claim\s*Status|Status)\s*[:\-]?\s*([A-Za-z ]+)",1
    ).alias("claim_status")
)

display(prop_df)
```

```python
claim_combined = (
    auto_df
    .unionByName(health_df, allowMissingColumns=True)
    .unionByName(prop_df,  allowMissingColumns=True)
)

display(claim_combined)
```

```python
def to_num(col_name):
    return F.when(
        F.col(col_name) == "", None   # convert empty → null
    ).otherwise(
        F.regexp_replace(F.col(col_name), ",", "").cast("double")
    )
```

```python
# Convert string columns to numeric, calculate totals, validate consistency, and enrich claim records
claim_gold = (
    claim_combined

    # Convert extracted string fields to numeric (double), handling empty values as null
    .withColumn("vehicle_repair",    to_num("vehicle_repair"))
    .withColumn("spare_parts",       to_num("spare_parts"))
    .withColumn("towing_charges",    to_num("towing_charges"))
    .withColumn("survey_fee",        to_num("survey_fee"))
    .withColumn("room_charges",      to_num("room_charges"))
    .withColumn("ot_fee",            to_num("ot_fee"))
    .withColumn("anaesthesia",       to_num("anaesthesia"))
    .withColumn("medicines",         to_num("medicines"))
    .withColumn("diagnostic_tests",  to_num("diagnostic_tests"))
    .withColumn("icu_charges",       to_num("icu_charges"))
    .withColumn("structural_damage", to_num("structural_damage"))
    .withColumn("machinery_damage",  to_num("machinery_damage"))
    .withColumn("stock_loss",        to_num("stock_loss"))
    .withColumn("electrical",        to_num("electrical"))
    .withColumn("debris_removal",    to_num("debris_removal"))
    .withColumn("business_loss",     to_num("business_loss"))
    .withColumn("total_claim_amount",to_num("total_claim_amount"))

    # Calculate total from extracted fields based on claim type
    .withColumn("calculated_total",
        F.when(F.col("claim_type") == "AUTO",
            F.coalesce(F.col("vehicle_repair"),    F.lit(0.0)) +
            F.coalesce(F.col("spare_parts"),       F.lit(0.0)) +
            F.coalesce(F.col("towing_charges"),    F.lit(0.0)) +
            F.coalesce(F.col("survey_fee"),        F.lit(0.0))
        )
        .when(F.col("claim_type") == "HLTH",
            F.coalesce(F.col("room_charges"),      F.lit(0.0)) +
            F.coalesce(F.col("ot_fee"),            F.lit(0.0)) +
            F.coalesce(F.col("anaesthesia"),       F.lit(0.0)) +
            F.coalesce(F.col("medicines"),         F.lit(0.0)) +
            F.coalesce(F.col("diagnostic_tests"),  F.lit(0.0)) +
            F.coalesce(F.col("icu_charges"),       F.lit(0.0))
        )
        .when(F.col("claim_type") == "PROP",
            F.coalesce(F.col("structural_damage"), F.lit(0.0)) +
            F.coalesce(F.col("machinery_damage"),  F.lit(0.0)) +
            F.coalesce(F.col("stock_loss"),        F.lit(0.0)) +
            F.coalesce(F.col("electrical"),        F.lit(0.0)) +
            F.coalesce(F.col("debris_removal"),    F.lit(0.0)) +
            F.coalesce(F.col("business_loss"),     F.lit(0.0))
        )
    )

    # Validate if calculated total matches extracted total_claim_amount (within tolerance)
    .withColumn("diff",
        F.abs(F.col("total_claim_amount") - F.col("calculated_total")))
    .withColumn("is_consistent",
        F.col("diff") <= F.lit(1.0))

    # Assign status flag based on extraction and consistency checks
    .withColumn("status",
        F.when(F.col("total_claim_amount").isNull(), "MISSING_TOTAL")
         .when(F.col("calculated_total").isNull(),   "EXTRACTION_FAILED")
         .when(~F.col("is_consistent"), "TOTAL_MISMATCH")
         .otherwise("OK")
    )

    # Map claim_type to descriptive label
    .withColumn("claim_type_full",
        F.when(F.col("claim_type") == "AUTO", "Auto Accident")
         .when(F.col("claim_type") == "HLTH", "Health / Medical")
         .when(F.col("claim_type") == "PROP", "Property Damage")
    )

    # Extract file name from path
    .withColumn("file_name",
        F.regexp_extract(F.col("path"), r"([^/]+\.pdf)$", 1))
)

display(claim_gold)
```
